```python
# ============================================
# Data Cleaning and Basic Data Analysis
# Dataset: Titanic
# ============================================
# Import the libraries that we'll use throughout the notebook.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Display plots inside the notebook
%matplotlib inline
# Use a clean style for the graphs
sns.set_style("whitegrid")
```

```python
# Load the Titanic dataset using Pandas.
# The CSV file should be in the same folder as this notebook.

df = pd.read_csv("Titanic.zip")

# Display the first five rows
df.head()
```

```python
# Display basic information about the dataset

print("Dataset Shape:")
print(df.shape)

print("\nDataset Information:")
df.info()

print("\nData Types:")
print(df.dtypes)

print("\nFirst Five Rows:")
display(df.head())
```

```python
# Check missing values
print("Missing Values:")
print(df.isnull().sum())

# Check duplicate rows
print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
# Fill missing values in Age using the median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing values in Embarked using the mode
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Drop the Cabin column because most of its values are missing
df.drop(columns=["Cabin"], inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

print("Dataset cleaned successfully.")
```

```python
# Summary statistics
print(df.describe())

# Mean Age
print("\nMean Age:")
print(df["Age"].mean())

# Median Fare
print("\nMedian Fare:")
print(df["Fare"].median())

# Number of passengers by gender
print("\nGender Counts:")
print(df["Sex"].value_counts())
```

```python
#Histogram
plt.figure(figsize=(8,5))

df["Age"].hist(bins=20, edgecolor="black")

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
#Bar Chart
plt.figure(figsize=(7,5))

df["Sex"].value_counts().plot(kind="bar")

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```
